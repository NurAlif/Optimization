```python
import plotly.express as px
import pandas as pd
import numpy as np


n_points = 200  

from pymoo.util.ref_dirs import get_reference_directions
ref_dirs = get_reference_directions("das-dennis", 3, n_partitions=12)


df = pd.DataFrame(new_data_2, columns=['V', 'W', 'X', 'Y'])
df['WV_combined'] = df['W'] + df['V']

from pymoo.util.nds.non_dominated_sorting import NonDominatedSorting

nds = NonDominatedSorting()
I = nds.do(df.F, only_non_dominated_front=True)
pareto_front = df.F[I]

df["is_pareto"] = False 
df.iloc[I, df.columns.get_loc("is_pareto")] = True 

fig = px.scatter_3d(df, x='X', y='Y', z='WV_combined', color='is_pareto',  
                    opacity=0.5,           
                    title='3D Pareto Front (WV combined, X and Y separate)',
                    labels={'is_pareto': 'Pareto Front'}, 
                    color_discrete_map={False: 'blue', True: 'red'}) 


fig.show()
```
